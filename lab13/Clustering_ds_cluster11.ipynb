## Librerías

````python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Clustering por K-means
from sklearn.cluster import KMeans
# silueta
from yellowbrick.cluster import SilhouetteVisualizer
# Clustering aglomerativo
from sklearn.cluster import AgglomerativeClustering
# Clustering con DBScan
from sklearn.cluster import DBSCAN
````

## Importación de datos

````python
#lectura de datos
datos = pd.read_csv("./ds_cluster11.csv", sep=',')
datos.head()
````

````python
# Convierte datos del dataframe a array de numpy
X = datos.values
````

````python
plt.scatter(datos.x, datos.y, s=8, c='black')
plt.grid(False)
````

# **1. Clustering con K-Means**

## **1.1. Aplicación de k-means, para k=2**

#### Entrenamiento

````python
# k-means con k=2
modelo_KAMEANS = KMeans(n_clusters=2)
modelo_KAMEANS.fit(X)
````

#### Predicción

````python
pred = modelo_KAMEANS.predict(X)
pred
````

### Resultado de agrupamiento con k-means, k= 2

````python
plt.scatter(X[:, 0], X[:, 1], c=pred, s=8, cmap='brg')
plt.grid(False)
````

#### Obtención de la silueta

````python
modelo_silueta = SilhouetteVisualizer(modelo_KAMEANS, colors='yellowbrick')

modelo_silueta.fit(X)
modelo_silueta.show()
plt.show()
print('Silueta = ', modelo_silueta.silhouette_score_)
````

## **1.2. Aplicación de k-means, para k=3**

#### Entrenamiento

````python
# k-means con k=3
modelo_KAMEANS = KMeans(n_clusters=3)
modelo_KAMEANS.fit(X)
````

#### Predicción

````python
pred = modelo_KAMEANS.predict(X)
pred
````

#### Resultado de agrupamiento con k-means, k=3

````python
plt.scatter(X[:, 0], X[:, 1], c=pred, s=8, cmap='brg')
plt.grid(False)
````

# **2. Clustering Jerarquico Aglomerativo**

### Entrenamiento

````python
modelo_CJA = AgglomerativeClustering(n_clusters=3)
modelo_CJA.fit(X)
````

### Predicción

````python
pred = modelo_CJA.labels_
pred
````

### Resultado de agrupamiento con CJA

````python
plt.scatter(X[:, 0], X[:, 1], c=pred, s=8, cmap='brg')
plt.grid(False)
````

### Graficar dendograma

Para graficar el dendograma, se debe aplicar clustering jerárquico sin indicar un limite a la cantidad de clusters

````python
# n_clusters: Para graficar el dendograma, el número de clústeres debe tomar el valor de None
# distance_threshold: El umbral de distancia de vinculación por encima del cual los clústeres no se fusionarán.
modelo_CJA = AgglomerativeClustering(n_clusters=None, distance_threshold=0)
modelo_CJA.fit(X)
````

### función de grafico

````python
# Implementación de gráfico de dendograma de Dr. Cesar Beltran
from scipy.cluster.hierarchy import dendrogram

def plot_dendrogram(model, **kwargs):
  counts = np.zeros(model.children_.shape[0])
  n_samples = len(model.labels_)
  for i, merge in enumerate(model.children_):
    current_count = 0
    for child_idx in merge:
      if child_idx < n_samples:
        current_count += 1
      else:
        current_count += counts[child_idx - n_samples]
    counts[i] = current_count
  linkage_matrix = np.column_stack([model.children_, model.distances_, counts]).astype(float)
  dendrogram(linkage_matrix, **kwargs)
````

````python
plt.figure(figsize=(12, 6))
#top_n_lebels_of_the_dendogram: permite realizar la poda en algun nivel del dendograma
top_n_levels_of_the_dendogram = 3
plot_dendrogram(modelo_CJA, truncate_mode='level', p=top_n_levels_of_the_dendogram)
plt.title("Dendograma de Clustering Jerarquico Aglomerativo")
plt.xlabel("Indice (nuemro de puntos)")
plt.show()
````
