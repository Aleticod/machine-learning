### Implementacion de un modulo que convierte un atributo categorico a numerico utilizando las funciones **astype('category')** y **cat.codes**

```python
def Categorico_a_numerico(atributo):
    # Convertir el atributo al tipo categorico
    atributo = atributo.astype('category')
    # Convertir el atributo categorico a numerico
    return atributo.astype('category').cat.codes
```

## Convertir los atributos

## **3. NORMALIZACION DE DATOS**

## Implementar la normalizacion de datos por amplitud y distribucion

```python
# Transformacion por amplitud [0..1]
def Normalizacion_Amplitud(atributo):
    return (atributo - atributo.min()) / (atributo.max() - atributo.min())
```

```python
# Transformacion por distribucion [-1...1]
def Normalizacion_Distribucion(atributo):
    return (atributo - atributo.mean()) / (atributo.var())
```

## Normalizar todos los datos por distribucion

```python
# Atributos del datset
df_riesto_credito.columns
```

```python
# Normalizacion de datos por distribucion
df_riesto_credito['status'] = Normalizacion_Amplitud(df_riesto_credito['status'])
df_riesto_credito['duration'] = Normalizacion_Amplitud(df_riesto_credito['duration'])
df_riesto_credito['credit_history'] = Normalizacion_Amplitud(df_riesto_credito['credit_history'])
df_riesto_credito['purpose'] = Normalizacion_Amplitud(df_riesto_credito['purpose'])
df_riesto_credito['amount'] = Normalizacion_Amplitud(df_riesto_credito['amount'])
df_riesto_credito['savings'] = Normalizacion_Amplitud(df_riesto_credito['savings'])
df_riesto_credito['employment_duration'] = Normalizacion_Amplitud(df_riesto_credito['employment_duration'])
df_riesto_credito['installment_rate'] = Normalizacion_Amplitud(df_riesto_credito['installment_rate'])
df_riesto_credito['personal_status_sex'] = Normalizacion_Amplitud(df_riesto_credito['personal_status_sex'])
df_riesto_credito['other_debtors'] = Normalizacion_Amplitud(df_riesto_credito['other_debtors'])
df_riesto_credito['present_residence'] = Normalizacion_Amplitud(df_riesto_credito['present_residence'])
df_riesto_credito['property'] = Normalizacion_Amplitud(df_riesto_credito['property'])
df_riesto_credito['age'] = Normalizacion_Amplitud(df_riesto_credito['age'])
df_riesto_credito['other_installment_plans'] = Normalizacion_Amplitud(df_riesto_credito['other_installment_plans'])
df_riesto_credito['housing'] = Normalizacion_Amplitud(df_riesto_credito['housing'])
df_riesto_credito['number_credits'] = Normalizacion_Amplitud(df_riesto_credito['number_credits'])
df_riesto_credito['job'] = Normalizacion_Amplitud(df_riesto_credito['job'])
df_riesto_credito['people_liable'] = Normalizacion_Amplitud(df_riesto_credito['people_liable'])
df_riesto_credito['telephone'] = Normalizacion_Amplitud(df_riesto_credito['telephone'])
df_riesto_credito['foreign_worker'] = Normalizacion_Amplitud(df_riesto_credito['foreign_worker'])
df_riesto_credito['credit_risk'] = Normalizacion_Amplitud(df_riesto_credito['credit_risk'])
```
